```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from datetime import datetime
import matplotlib.dates as mdates
from matplotlib.pyplot import figure

%matplotlib inline
```

```python
## reset parameters
plt.rcParams.update(plt.rcParamsDefault)

plt.style.use('dark_background')
```

```python
font = {'family' : 'Segoe UI',
        'size'   : 12}

matplotlib.rc('font', **font)
```

```python
df = pd.read_excel("C:/Users/MCUSA88/Desktop/For Preston/covid19_onedrive/covid19_data.xlsx", sheet_name="whatcom_covid19_data")
```

```python
df.head()
```

```python
# segoeUIfont = {'fontname':'Segoe UI'}
```

```python
# fig= plt.figure(figsize=(6, 4))

dates = ['3/10/2020', '4/1/2020', '5/1/2020', '6/1/2020', '7/1/2020']
date_ticks = [datetime.strptime(d,"%m/%d/%Y").date() for d in dates]

cumulative_cases = plt.plot(df.date, df.cases, color='#1F1FFF', linewidth=2.5)

ax = plt.gca()
formatter = mdates.DateFormatter("%b %d")
ax.xaxis.set_major_formatter(formatter)
plt.xticks(ticks=date_ticks, rotation=0)

plt.title('CUMULATIVE CASES', size=18, pad=10)

plt.grid(True, linestyle='--', linewidth=0.25)

plt.savefig('cumulative_cases.png', bbox_inches='tight')

plt.show()

# plt.yticks(**segoeUIfont)
```

```python
dates = ['3/10/2020', '4/1/2020', '5/1/2020', '6/1/2020', '7/1/2020']
date_ticks = [datetime.strptime(d,"%m/%d/%Y").date() for d in dates]

cumulative_deaths = plt.plot(df.date, df.deaths, color='#FF0A0A', linewidth=2.5)

ax = plt.gca()
formatter = mdates.DateFormatter("%b %d")
ax.xaxis.set_major_formatter(formatter)

plt.xticks(ticks=date_ticks, rotation=0)
plt.title('CUMULATIVE DEATHS', size=18, pad=10)
plt.grid(True, linestyle='--', linewidth=0.25)

plt.savefig('cumulative_deaths.png', bbox_inches='tight')
plt.show()

# plt.yticks(**segoeUIfont)
```
